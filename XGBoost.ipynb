```python
X_encoded=pd.get_dummies(X,columns=['Phone_Service'
,'Multiple_Lines'
,'Internet_Service'
,'Online_Security'
,'Online_Backup'
,'Device_Protection'
,'Tech_Support'
,'Streaming_TV'
,'Streaming_Movies'
,'Contract'
,'Paperless_Billing'
 ,'Gender'
,'Senior_Citizen'
,'Partner'
,'Dependents','Payment_Method'
,'City'])
```

```python
clf_xgb=Xgb.XGBClassifier(objective='binary:logistic',missing=None,seed=42)
```

```python
clf_xgb.fit(X_train,
            y_train,
            verbose=True,
            early_stopping_rounds=10,
            eval_metric='aucpr',
            eval_set=[(X_test,y_test)])
```
